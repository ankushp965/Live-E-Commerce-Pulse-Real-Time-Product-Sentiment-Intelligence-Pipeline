```python
# PRODUCT_SEARCH: Real-time Amazon Electronics Data Fetcher
import requests
import pandas as pd
import time

# Initialize storage
all_products = []

# API endpoint and headers
url = "https://real-time-amazon-data.p.rapidapi.com/search"
headers = {
    "x-rapidapi-key": "9d012ee951mshc446ce71612c376p12ed21jsnd7492a13af9f",
    "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
}

# Fetch 70 pages of product data
for i in range(1, 71):
    querystring = {
        "query": "Electronics",
        "page": str(i),
        "country": "US",
        "sort_by": "RELEVANCE",
        "product_condition": "ALL",
        "is_prime": "false",
        "deals_and_discounts": "NONE"
    }

    try:
        response = requests.get(url, headers=headers, params=querystring)

        if response.status_code != 200:
            print(f"Error on page {i}: Status Code {response.status_code}")
            break  # Stop if something goes wrong

        data = response.json()
        products = data.get('data', {}).get('products', [])

        all_products.extend(products)
        print(f"Page {i} fetched, total so far: {len(all_products)}")

        time.sleep(1)  # Sleep to avoid hitting rate limits

    except Exception as e:
        print(f"Exception on page {i}: {e}")
        break

# Convert to DataFrame
df = pd.DataFrame(all_products)

# Save to CSV for ADLS upload
df.to_csv("amazon_products.csv", index=False)
print(f"\n Final Dataset Saved: {len(df)} records in 'amazon_products.csv'")
```
